# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
# Your code here
data = []
data_n = []

for col in ['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']:
    q25 = customers[col].quantile(q = 0.25)
    q75 = customers[col].quantile(q = 0.75)
    iqr = q75-q25
    data.append(customers[col]> q75+(iqr*1.5))
    data_n.append(customers[col]<= q75+(iqr*1.5))

print(customers[data[0]|data[1]|data[2]|data[3]|data[4]|data[5]].shape[0]/customers.shape[0])
# too much information is removed with outliers (25% of rows)
# I prefer to scale the data and keep those values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
cols = ['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']

customers_scaled = customers.copy()

scaler = StandardScaler()
scaler.fit(customers_scaled[cols])
arr = scaler.transform(customers_scaled[cols])

for i,col in enumerate(cols):
    customers_scaled[col] = arr[:,i]

customers_scaled
```

```python
customers_scaled_encoded = pd.get_dummies(data = customers_scaled, columns = ['Channel','Region'], drop_first=True)
customers_scaled_encoded.head()
```

# Challenge 3.1 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans
```

```python
# Your code here:
kmeans = KMeans(n_clusters = 4)
cust_clustered = kmeans.fit(customers_scaled_encoded)
customers['labels'] = cust_clustered.labels_
```

```python
# visually...
from sklearn.decomposition import PCA
pca = PCA(n_components = 2).fit_transform(customers_scaled_encoded)

plt.figure(figsize=(12,12))
plt.scatter(pca[:,0], pca[:,1], c = cust_clustered.labels_);
```

Count the values in `labels`.

```python
# Your code here:
customers['labels'].value_counts()
```

# Challenge 4 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN
# Your code here
dbscan = DBSCAN(eps = 0.5)
dbscan.fit(customers_scaled_encoded)
customers['labels_DBSCAN'] = dbscan.labels_
```

```python
# visually...
from sklearn.decomposition import PCA
pca = PCA(n_components = 2).fit_transform(customers_scaled_encoded)

plt.figure(figsize=(12,12))
plt.scatter(pca[:,0], pca[:,1], c = dbscan.labels_);
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here
fig, axs = plt.subplots(2, 4, figsize = (17,10))
axs = axs.flatten()

for i,ax in enumerate(axs):
    kmeans = KMeans(n_clusters = i+2)
    cust_clustered = kmeans.fit(customers_scaled_encoded)

    pca = PCA(n_components = 2).fit_transform(customers_scaled_encoded)
    
    ax.scatter(pca[:,0], pca[:,1], c = cust_clustered.labels_)
    ax.set_title(f'{i+2} Clusters')

plt.show();
```

```python
# Your comment here
# I would say 4 or 5 are the best configurations
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here
fig, axs = plt.subplots(4, 4, figsize = (17,10))
axs = axs.flatten()

for i,ax in enumerate(axs):
    dbscan = DBSCAN(eps = ((i//4)+1)/2, min_samples=3+(i%4)*2)
    dbscan.fit(customers_scaled_encoded)
    
    pca = PCA(n_components = 2).fit_transform(customers_scaled_encoded)
    
    ax.scatter(pca[:,0], pca[:,1], c = cust_clustered.labels_)
    ax.set_title(f'eps = {((i//4)+1)/2}, min_samp = {3+(i%4)*2}')

plt.show();
```
