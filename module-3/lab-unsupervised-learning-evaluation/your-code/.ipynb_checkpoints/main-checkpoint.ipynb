```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 8)
kmeans.fit(X)
y_pred = kmeans.labels_
print(silhouette_score(X, y_pred))
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 5)
kmeans.fit(X)
y_pred = kmeans.labels_
print(silhouette_score(X, y_pred))
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
y_pred = kmeans.labels_
print(silhouette_score(X, y_pred))
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer


visualizer = KElbowVisualizer(KMeans(), k=(2,12), metric='distortion')
visualizer.fit(X)
visualizer.poof();

visualizer = KElbowVisualizer(KMeans(), k=(2,12), metric='silhouette')
visualizer.fit(X)
visualizer.poof();

visualizer = KElbowVisualizer(KMeans(), k=(2,12), metric='calinski_harabasz')
visualizer.fit(X)
visualizer.poof();
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
visualizer = KElbowVisualizer(AgglomerativeClustering(), k=(2,12), metric='distortion')
visualizer.fit(X)
visualizer.poof();

visualizer = KElbowVisualizer(AgglomerativeClustering(), k=(2,12), metric='silhouette')
visualizer.fit(X)
visualizer.poof();

visualizer = KElbowVisualizer(AgglomerativeClustering(), k=(2,12), metric='calinski_harabasz')
visualizer.fit(X)
visualizer.poof();
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
import seaborn as sns
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(X)
X['label_kmeans'] = kmeans.labels_
```

```python
sns.pairplot(X, hue='label_kmeans');
```

```python
X.drop(columns='label_kmeans', inplace=True)
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(X)
pca_X = pca.transform(X)
```

```python
# sns.scatterplot(pca_X[:,0], pca_X[:,1], hue = kmeans.labels_);

LABEL_COLOR_MAP = {0:'r',1:'b',2:'g',3:'k'}
label_color = [LABEL_COLOR_MAP[l] for l in kmeans.labels_]

plt.scatter(pca_X[:,0], pca_X[:,1], c=label_color);
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE
```

```python
fig, axs = plt.subplots(3,3,figsize = (12,12))
axs = axs.flatten()

LABEL_COLOR_MAP = {0:'r',1:'b',2:'g',3:'k'}
label_color = [LABEL_COLOR_MAP[l] for l in kmeans.labels_]


for i,ax in enumerate(axs):
    X_emb = TSNE(n_components=2, perplexity=(5+i*3)).fit_transform(X)
        
    
    ax.scatter(X_emb[:,0], X_emb[:,1], c=label_color);
    ax.set_title(f'perplexity: {5+i*3}')

plt.show();
```
